####  Display the 20 latest earthquakes info (date, time, latitude, longitude and region name) by the EMSC as a pandas dataframe.

````python
# This is the url you will scrape in this exercise
url_earthquake = 'https://www.emsc-csem.org/Earthquake/'
````

````python
# your code here
url_earthquake = 'https://www.emsc-csem.org/Earthquake/'
html_earthquake = requests.get(url_earthquake).content
soup_earthquake = BeautifulSoup(html_earthquake, "html")
earthquakes_lines = soup_earthquake.find_all('tr',{"class":["ligne1 normal","ligne2 normal"]})
earthquakes_info = []
for earthquake in earthquakes_lines:
    datetime = earthquake.find_all('td',{"class": "tabev6"})[0].text.split()
    date = (datetime[0])[-10:]
    time = " ".join(datetime[1:])[:8]
    latitude = f"{earthquake.find_all('td',{'class': 'tabev1'})[0].text.strip()} {earthquake.find_all('td',{'class': 'tabev2'})[0].text.strip()}"
    longitude = f"{earthquake.find_all('td',{'class': 'tabev1'})[1].text.strip()} {earthquake.find_all('td',{'class': 'tabev2'})[1].text.strip()}"
    region = earthquake.find_all('td',{"class": "tb_region"})[0].text.strip()
    earthquakes_info.append({"date":date, "time": time, "latitude": latitude,"longitude": longitude,"region":region})
    
earthquakes_info[:20]
pd.DataFrame(earthquakes_info[:20])
````

#### Count the number of tweets by a given Twitter account.

#### List all language names and number of related articles in the order they appear in wikipedia.org.

````python
# This is the url you will scrape in this exercise
url_wikipedia = 'https://www.wikipedia.org/'
````

````python
# your code here
html_wikipedia = requests.get(url_wikipedia).content
soup_wikipedia = BeautifulSoup(html_wikipedia, "html")

res = []
for language in soup_wikipedia.find_all('div', {'class':'central-featured-lang'}):
    title = language.a.strong.text
    # article = language.a.small.text
    article = "".join(filter(str.isdigit, language.a.small.text))
    res.append([title, article])
    
res_df = pd.DataFrame(res)
res_df
````
