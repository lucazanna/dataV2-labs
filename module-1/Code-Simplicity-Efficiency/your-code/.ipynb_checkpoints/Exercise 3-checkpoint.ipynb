```python
import random
```

```python
def my_function(X):
    solutions = []
    for x in range(5, X):
        for y in range(4, X):
            for z in range(3, X):
                if (x*x==y*y+z*z):
                    solutions.append([x, y, z])
    m = 0
    for solution in solutions:
        if m < max(solution):
            m = max(solution)
    return m

X = input("What is the maximal length of the triangle side? Enter a number: ")

print("The longest side possible is " + str(my_function(int(X))))
```

```python
import math
```

```python
def my_function_LZ(X):

    side = 0;
    check = False;
    
    for x in list(range(5, X))[::-1]:
        for y in range(4, x):
            z = math.sqrt((x**2-y**2))
            if z.is_integer() and z!= 0:
                side = x;
                check = True;
        if check:
            break;
#    print(f"side {side}")
    return side
```

```python
my_function_LZ(20)
```

```python
[(my_function_LZ(number),my_function(number)) for number in random.sample(range(5, 100), 30)]
```
