```python
import random
import string
```

```python
# generates a random string of l characters with lowercase letters and digits
def random_string_generator(l=12):
    return "".join([random.choice(list(string.ascii_lowercase) + list(string.digits)) for character in range(l)])
```

```python
# generates a list of n strings, between a and b characters
def BatchStringGenerator(n, a=8, b=12):
    if b<a:
        print("Incorrect min and max string lengths. Try again.")
    else:
        return [ random_string_generator(random.choice(range(a, b+1))) for number in range(n)]
```
