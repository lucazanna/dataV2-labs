```python
# exercise 2 v1
class List(object):
            
    def count_spec_digits(self, integers_list, digits_list):
        # your code here
        return [(number,  "".join(map(str,integers_list)).count(str(number)) ) for number in digits_list]
```

```python
# exercise 2 v2
class List(object):
    def __init__ (self, integers_list):
        self.str_list = "".join(map(str,integers_list))
        self.counts = [(number, self.str_list.count(str(number))) for number in range(10)]
    def count_spec_digits(self, digits_list):
        # your code here
        return list(filter(lambda x: x[0] in digits_list, self.counts))
```
